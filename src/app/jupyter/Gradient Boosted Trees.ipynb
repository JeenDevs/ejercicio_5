```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

#dataframe = pd.read_csv('Prueba2.csv', sep=';',encoding ='latin-1')
#print(dataframe)
dataframe = pd.read_csv('Datos.csv', sep=';',encoding ='latin-1')
print(dataframe)
```

```python
import matplotlib.pyplot as plt

dataframe.dropna(inplace=True)
dataframe.isnull().sum().sum()
X = dataframe.iloc[:, 2:7].values
y = dataframe.iloc[:, 7].values
print(dataframe)
# plt.plot(dataframe)
# plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# plt.plot(train_test_split)
# plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_test)
plt.plot(X_test)
plt.show()
```

```python

import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor

gb_reg = GradientBoostingRegressor(n_estimators = 100)
gb_reg = gb_reg.fit(X_train, y_train)

# generating predictions for the gradient boosting regressor
y_pred_gb = gb_reg.predict(X_test)
plt.plot(y_pred_gb)
plt.show()
```

```python
# mean squared error
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
mse_gb = np.round(mean_squared_error(y_test, y_pred_gb), 2)
mse_gb
# plt.plot(mse_gb)
# plt.show()
```

```python
# r2 score
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
r2_score_gb = np.round(r2_score(y_test, y_pred_gb), 3)
r2_score_gb
# plt.plot(r2_score_gb)
# plt.show()
```
